```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

Change the filepath if you would like to graph a different CSV

```python
# mnist
# filename = '/home/yoshisada/Desktop/condensa/notebooks/condensa_outputs/master_pruning_data.csv'

# cifar
filename = '/home/yoshisada/Desktop/condensa/notebooks/cifar10_outputs/master_pruning_data.csv'
```

```python
# mnist
# OUT_PATH = '/home/yoshisada/Desktop/condensa/notebooks/condensa_graphs'

# cifar
OUT_PATH =  '/home/yoshisada/Desktop/condensa/notebooks/cifar10_graphs'
```

```python
now = datetime.now()
# mm_dd_YY_H:M:S
dt_string = now.strftime("%m-%d-%Y_%H:%M:%S")
print("date and time =", dt_string)
```

```python
# first read it in
df = pd.read_csv(filename)
```

```python
df
```

```python
df.plot(x='PruningDensity', title='Pruning Density VS Accuracy', y=['Accuracy'], figsize=(10,6))
plt.gca().invert_xaxis()
plt.savefig(f'{OUT_PATH}/density_v_acc/den_v_acc_{dt_string}.png')
```

```python
df.plot(x='PruningDensity', title='Pruning Density VS Non Zero Values', y=['NonZeros'], figsize=(10,6))
plt.gca().invert_xaxis()
plt.savefig(f'{OUT_PATH}/density_v_nonZeros/den_v_nonZero_{dt_string}.png')
```

'%%script false --no-raise-error' skips the cell below if you would like to run it, comment it out and then run the cell

```python
# %%script false --no-raise-error
# graph density vs compressionTime

# Divide all the y values by 60 to get time in mins
df['CompressionTime'] = df['CompressionTime'] / 60

df.plot(x='PruningDensity', title='Pruning Density VS Compression Time (in mins)', y=['CompressionTime'], figsize=(10,6))
plt.gca().invert_xaxis()
plt.savefig(f'{OUT_PATH}/density_v_compTime/den_v_compTime_{dt_string}.png')
```
